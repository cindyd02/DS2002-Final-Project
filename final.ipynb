# ETL Implementation

```python
# filter the vaccinations dataset for US data
us_vaccinations = vaccinations[(vaccinations['country'] == 'United States') | (vaccinations['iso_code'] == 'USA')]

print('\nFiltered Vaccinations Data for United States: \n')
us_vaccinations.head()
```

```python
# checking for missing values
print("\nMissing values in Vaccinations dataset:")
print(us_vaccinations.isnull().sum())

print("\nMissing values in Records dataset:")
print(records.isnull().sum())
```

```python
# remove rows with missing values
us_vaccinations_cleaned = us_vaccinations.dropna()

# check for missing values in the cleaned dataset
print("\nMissing values in cleaned US Vaccinations dataset:")
print(us_vaccinations_cleaned.isnull().sum())

print(f"\nRows before cleaning: {len(us_vaccinations)}")
print(f"Rows after cleaning: {len(us_vaccinations_cleaned)}")
```

```python
import numpy as np
# dropping infinite values in records dataset
print("\nInfinite values in records:")
print(records.isin([np.inf, -np.inf]).sum())
records.replace([np.inf, -np.inf], np.nan, inplace=True)

print("\nInfinite values in records afterwards:")
print(records.isin([np.inf, -np.inf]).sum())

```

```python
# Load Data into MySQL
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

# Load environment variables from .env
load_dotenv()

# Define the connection string to connect to MySQL
connection_string = os.getenv('DB_CONNECTION_STRING')

# Create the connection engine using the connection string
engine = create_engine(connection_string)

# Loading the us_vaccinations_cleaned dataset into MySQL
us_vaccinations_cleaned.to_sql('vaccination_data', con=engine, if_exists='replace', index=False)

# Loading records dataset into MYSQL
records.to_sql('records_data', con=engine, if_exists='replace', index=False)

print("Succesful! Data uploaded to MySQL.")
```

## Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
vaccinations = pd.read_csv('country_vaccinations.csv')
records = pd.read_csv('country_wise_latest.csv')

us_vaccinations = vaccinations[vaccinations['country'] == 'United States'].copy()
us_vaccinations['date'] = pd.to_datetime(us_vaccinations['date'])

# analyze vaccinations data
plt.figure(figsize=(12, 6))
plt.plot(us_vaccinations['date'], us_vaccinations['total_vaccinations'])
plt.title('Total Vaccinations in the US Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(us_vaccinations['date'], us_vaccinations['daily_vaccinations'])
plt.title('Daily Vaccinations in the US')
plt.xlabel('Date')
plt.ylabel('Daily Vaccinations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# analyze records data
us_records = records[records['Country/Region'] == 'US']
metrics = ['Confirmed', 'Deaths', 'Recovered', 'Active']

plt.figure(figsize=(12, 8))
for metric in metrics:
    plt.bar(metric, us_records[metric].values[0])
plt.title('COVID-19 Cases in the US')
plt.ylabel('Number of Cases')
plt.tight_layout()
plt.show()

# summary statistics
print("Vaccination Data Summary:")
print(us_vaccinations[['total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'daily_vaccinations']].describe())

print("\nCOVID-19 Cases Summary:")
print(us_records[metrics].T)

# additional analysis: vaccination rate over time
plt.figure(figsize=(12, 6))
plt.plot(us_vaccinations['date'], us_vaccinations['people_vaccinated_per_hundred'])
plt.title('Percentage of Population Vaccinated in the US')
plt.xlabel('Date')
plt.ylabel('People Vaccinated per Hundred')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# distribution of daily vaccinations
plt.figure(figsize=(10, 6))
sns.histplot(us_vaccinations['daily_vaccinations'], kde=True)
plt.title('Distribution of Daily Vaccinations')
plt.xlabel('Daily Vaccinations')
plt.show()
```

## Summary of findings supported by the visualizations above

- Daily rates show significant variation, with a standard deviation of about 804,436 doses (8.044360e+05).
- The highest recorded daily vaccination rate was 3.51 million doses (3.506960e+06).
3. Vaccination Distribution
- The distribution of daily vaccinations likely shows a right-skewed pattern, given the difference between the mean (1.19 million) and median (960,812) daily doses.
- This suggests more frequent occurrences of lower daily vaccination numbers, with occasional high-volume days pulling the average up.
4. COVID-19 Cases in the US
- Confirmed Cases: 4,290,259
- Deaths: 148,011
- Recovered: 1,325,804
- Active Cases: 2,816,444
5. Vaccination Coverage
- The percentage of the population vaccinated has likely increased over time, as shown in the "Percentage of Population Vaccinated in the US" graph.
- The maximum percentage of people who received at least one dose is about 77.5% (255.3 million out of an estimated US population of 329.5 million in 2020).
6. Relationship Between Vaccinations and Cases
- While we can't draw direct correlations without time-series data, the high number of vaccinations compared to active cases (2.8 million) suggests that vaccinations may have played a role in controlling the spread of COVID-19.
7. Vaccination Efficiency
- The difference between total vaccinations (560.2 million max) and fully vaccinated individuals (217.5 million max) indicates that a significant portion of the population has received multi-dose vaccines.

This summary provides an overview of the vaccination efforts and COVID-19 situation in the United States. The high vaccination numbers, particularly in total doses and people receiving at least one dose, suggest a substantial vaccination campaign. However, the presence of active cases indicates ongoing challenges in fully controlling the spread of the virus. The variability in daily vaccination rates might reflect logistical challenges or changes in vaccine availability and public response over time.
